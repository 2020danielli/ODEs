```python
import numpy as np
import matplotlib.pyplot as plt

def curves(a, b, l, T):
    t = np.linspace(0, T, 1000)

    x = np.cos(a * t)
    y = np.sin(b * t)

    x_l = np.cos(l * a * t)
    y_l = np.sin(l * b * t)

    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.title(f"a = {a}, b = {b}")
    plt.plot(x, y)

    plt.subplot(122)
    plt.title(f"λ = {l}")
    plt.plot(x_l, y_l)

    plt.tight_layout()
    plt.show()

curves(1,(1 + np.sqrt(5)) / 2,2,50)
```

```python
curves(1,1,10,50)
```

```python
curves(1,3,10,50)
```

```python
curves(1,5,10,50)
```

```python
curves(5,1,10,50)
```

```python
def dist(f, N):
    bins = np.zeros(9)
    running = 1
    for n in range(1, N + 1):
        running *= f

        d = int(str(running)[0])

        if d != 0:
            bins[d - 1] += 1

    total = np.sum(bins)
    frequencies = bins/total

    plt.bar(range(1, 10), frequencies)
    plt.title(f"First digit frequency of {f} to up to the {N}th power")
    plt.xlabel("digit")
    plt.ylabel("freq")
    plt.xticks(np.arange(1, 10, 1))
    plt.show()
dist(2, 1000)
```

```python
dist(3,1000)
```
